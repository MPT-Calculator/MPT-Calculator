## Example 5 - Using Adaptive POD With a Step Geometry
In this example, we cover the use of the adaptive POD with geometry loaded in from a $\texttt{.step}$ file description. 

The $\texttt{.step}$ file describes a small cube of side length 10 mm and is located [here](./Examples/Example_10mm_cube.step). The corresponding OCC file is defined in [OCC_step_cube.py](../OCC_Geometry/OCC_step_cube.py). The OCC file sets $\mu_r=1$, $\sigma_*=10^6$ S/m, and $\alpha=0.001$ m and can be updated easily for new parameters or geometries.

## Usage
In this example, we wish to consider 80 logarithmically spaced frequencies between $1\times 10^1$ rad/s and $1\times 10^8$ rad/s using an initial 13 logarithmically spaced snapshots using an order $p=2$ approximation.

and specify the sweep parameters we are interested in. Note that $\alpha$ has been specified directly in the OCC file.

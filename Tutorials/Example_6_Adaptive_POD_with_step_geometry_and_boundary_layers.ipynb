## Example 6 - Using Adaptive POD With a Step Geometry and Boundary Layers
In this example, we cover the use of the adaptive POD with geometry loaded in from a $\texttt{.step}$ file description. 

The $\texttt{.step}$ file describes a small cube of side length 10 mm and is located [here](./Examples/Example_10mm_cube.step). The corresponding OCC file is defined in [OCC_step_cube_prism.py](../OCC_Geometry/OCC_step_cube_prism.py). The OCC file sets $\mu_r=100$, $\sigma_*=10^6$ S/m, and $\alpha=0.001$ m and can be updated easily for new parameters or geometries.

## Structure of Boundary Layers
Following the advice in Tutorial [7](./7_Boundary_layer_elements.ipynb), we assign 2 layers of prismatic elements of thickness $\tau$ and $2\tau$, with $\tau$ being the normalised skin depth. We also specify that we desire a homogeneous cube, so the material for the boundary layers matches the material for the cube interior. These settings can easily be updated in the OCC file.

Be aware that using large numbers of prismatic elements dramatically increases the number of degrees of freedom (and potentially the matrix condition number) and will typically take significantly longer to run than the using a pure tetrahedral mesh with the same settings.

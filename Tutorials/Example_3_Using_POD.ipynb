## Example 3 - Using POD
In this example, we demonstrate the use of the POD mode of MPT Calculator.
We consider the MPT characterisation of a conducting sphere with $\sigma_*=10^6$ S/m, $\mu_r=1$, and radius $\alpha=0.01$ m.

When using POD, the default settings are to have 13 full order model solutions obtained at logarithmically spaced frequency snapshots and to extract model information based on a singular value decomposition using a tolerance of $10^{−6}$. Reduced order model solutions are then computed at 80 logarithmically spaced frequency points between $10^1$ and $10^8$ rad/s.

The geometry for this problem is defined in [OCC_sphere.py](../OCC_Geometry/OCC_sphere.py) which can be easily edited for different materials.

First we set the working directory and import the main function.

```python
%load_ext autoreload
%autoreload 2

import os
# Please edit this line to provide the to the root directory of your copy of MPT-Calculator
Your_Path= r'C:\Users\James\Desktop\MPT-Calculator'
os.chdir(Your_Path)
os.getcwd()

from main import main
```

and set the the frequency sweep, geometry, and discretisation before running $\texttt{main()}$ with the $\texttt{use_POD}$ argument set to True.

```python
geometry = 'OCC_sphere.py'
start = 1
stop = 8
points = 80
order = 2
alpha = 1e-2
mesh_coarseness = 'coarse'

start_stop = (start, stop, points)

result = main(geometry=geometry, alpha=alpha, order=order, h=mesh_coarseness, start_stop=start_stop, use_OCC=True,
             use_POD=True)
```

p refinement example using MPT-Calculator

```python
from main import main
from time import time
import numpy as np
from matplotlib import pyplot as plt
```

Specifying meshsize h as a coarse mesh.

```python
h = 2
```

performing p refinement for orders 0 to 3 and comparing eigenvalues

```python
comparison_eig = np.zeros((40,4), dtype=complex)
for p in [0,1,2,3]:
    TensorArray, EigenValues, N0, elements, Array, ndofs, EddyCurrentTest = main(use_POD=True, hp=(h,p))
    comparison_eig[:,p] = EigenValues[:,0]
```

```python
for p in [0,1,2,3]:
    plt.semilogx(Array, comparison_eig[:,p].real, label=f'order {p}')
plt.legend();
plt.xlabel('$\omega$, [rad/s]');
plt.ylabel('$\lambda_1(\mathcal{R} + \mathcal{N}^0)$');
```

```python
for p in [0,1,2,3]:
    plt.semilogx(Array, comparison_eig[:,p].imag, label=f'order {p}')
plt.legend();
plt.xlabel('$\omega$, [rad/s]');
plt.ylabel('$\lambda_1(\mathcal{I})$');
```

# $hp$ Refinement example using MPT-Calculator

This `jupyter notebook` explores 2 different routes to improving the accuracy of full order solutions computed using `MPT-Calculator` namely

1. $p$-refinement: where a fixed mesh is used and the polynonial order of the elements is uniformly increased until convergence of the solution is achieved.

2. $h$-refinement: where the polynomial order is fixed and the mesh spacing is refined until convergence of the solution is achieved.

Generally, we will want to use $p$ refinement and, if necessary, construct a mesh with local $h$ refinement to capture small scale features.

## $p$ - Refinement

In this example, we wish to explore the effect of changing the polynomial order of the $\boldsymbol{H}(curl)$ conforming finite element discretisation. We expect that increasing the order $p$ from $p=0\rightarrow p=1\rightarrow p=2\rightarrow p=3$ will result in a more accurate simulation at the cost of an increased computational time. We want to ensure that we have chosen $p$ to be sufficently high so that the MPT coefficents (or their eigenvalues) are converged (i.e. further increases in $p$ do not substantially change their values).

For control over the order we want to specify $p$ as an input argument
```python
main(order=p)
```

which can be done iteratively. We apply this to the MPT characterisation of a conducting sphere with $\sigma_* =1  \times 10^6 $ S/m, $\mu_r=1$ and radius $\alpha =1 \times 10^{-3}$ m using the same default descretisation as in Tutorial ?? We loop through different orders as follows:

````python
comparison_eig = np.zeros((40,4), dtype=complex)
for p in [0,1,2,3]:
    TensorArray, EigenValues, N0, elements, Array, ndofs, EddyCurrentTest = main(use_POD=True, order=p)
    comparison_eig[:,p] = EigenValues[:,0]
````

Plotting the first eigenvalue for the different $p$, shows that the solutions are in agreement upto $\sim 10^6$ rad/s, after which the solutions begin to diverge.

````python
for p in [0,1,2,3]:
    plt.semilogx(Array, comparison_eig[:,p].real, label=f'order {p}')
plt.legend();
plt.grid();
plt.xlabel('$\omega$, [rad/s]');
plt.ylabel('$\lambda_1(\mathcal{R} + \mathcal{N}^0)$');
````

````python
for p in [0,1,2,3]:
    plt.semilogx(Array, comparison_eig[:,p].imag, label=f'order {p}')
plt.legend();
plt.grid();
plt.xlabel('$\omega$, [rad/s]');
plt.ylabel('$\lambda_1(\mathcal{I})$');
````

To achieve converged solutions for higher frequencies we can

1. Increase the order $p$ of the elements further and/or
2. Use a mesh with more elements
